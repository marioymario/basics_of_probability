#  Law of large numbers
* Related to the central limit theorem, the law of large numbers assures 

  us that as our sample size grows, the sample mean becomes very close to 

  the population mean.

# demonstration: law of large numbers
* For this demonstration, we will conduct 1,000 simulations of die throws. 

  At each simulation, we'll add one more throw to our sample, & take the average each time.

```python
results = []
for num_trials in range(1,1000):
    trials = np.random.randint(1, 7, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()
 
```

With these assumptions at hand, we're able to compare a given sample to the 

normal sampling distribution to assess the likelihood that our sample comes 

from the same population. In other words, we can test the hypothesis that 

our sample represents a population distinct from the known population. 

We can also compare for significant differences in means of two populations, 

given that each is reflective of their populations.
